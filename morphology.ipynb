# morphology transformation and image segmentation

# dilation

```python
import cv2 
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("C:/Users/SN127/tharun/pul.jpg",0)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
_,mask = cv2.threshold(img,220,255,cv2.THRESH_BINARY)
img
```

```python
kernel=np.ones((3,3),np.uint8)
dilation=cv2.dilate(mask,kernel,iterations=1)
titles=['image','mask','dilate']
images=[img,mask,dilation]
for i in range(len(images)):
    plt.subplot(1,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),    
```

# morphological opening

```python
open=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)

```

```python
cv2.imshow("image",img)
cv2.imshow('open',open)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# morphological closing followed by dilationand ersion

```python
close=cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('close',close)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# morphological gradient-erosion,dilation and difference of both

```python
gradient=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('morph gradient',gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# top hat-opening and diff of original and opening

```python
tophat=cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('tophat',tophat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# black hat 

```python
blackhat=cv2.morphologyEx(img,cv2.MORPH_BLACKHAT,kernel)
```

```python
cv2.imshow("image",img)
cv2.imshow('black hat',blackhat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
